```python
import numpy as np
def addopa(image):
    shape = image.shape
    size_x = shape[0]
    size_y = shape[1]
    size_channel = shape[2]
    pixel_x = np.random.randint(size_x)
    pixel_y = np.random.randint(size_y)
    pixel_channel = size_channel
    image[pixel_x,pixel_y] = np.random.uniform(0,1)
    return image


from PIL import Image
from numpy import asarray
# load the image
image = Image.open('00001_00043_00002_png_jpg.rf.94e580a9da3d2d679e73089ca5ba1244.jpg')
# convert image to numpy array
data = asarray(image)
image = addopa(data)
```
